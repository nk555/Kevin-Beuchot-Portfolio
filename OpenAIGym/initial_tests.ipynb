```python
import gym
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
env = gym.make('CartPole-v1')
i=1000
j=100
curr_t=0
for i_episode in range(i):
    observation = env.reset()
    tot_rew=0
    this_obs_act_rew=[]
    rang=j
    for t in range(j):
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        this_obs_act_rew.append(np.array(list(observation)+[action]+[0]))
        if done:
            rang=t
            break
    this_obs_act_rew=np.array(this_obs_act_rew)
    if i_episode==0:
        obs_act_rew=this_obs_act_rew
    else:
        obs_act_rew=np.concatenate([obs_act_rew,this_obs_act_rew], axis=0)
    for t in range(rang):
        if rang-t>=5:
            obs_act_rew[curr_t+t,5]=1
    curr_t=curr_t+rang
env.close()
```

```python
from tensorflow.keras import layers
model=tf.keras.Sequential(layers=[layers.Dense(50, activation="relu"), layers.Dense(100, activation="relu"), layers.Dense(50, activation="relu"), layers.Dense(2, activation="relu"), layers.Softmax()])
model.compile(optimizer="adam", loss="binary_crossentropy")
model.fit(obs_act_rew[:, 0:5], obs_act_rew[:, 5], epochs=100)
```

```python
observation = env.reset()
observation, reward, done, info = env.step(0)
while done!=True:
    env.render()
    best_action=0
    best_reward=0
    for action in range(2):
        pred_rew=model.predict(np.array([np.array(list(observation)+[action])]))
        if pred_rew>best_reward:
            best_reward=pred_rew
            best_action=action
    observation, reward, done, info = env.step(best_action)
env.close()
```
